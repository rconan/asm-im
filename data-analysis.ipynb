```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("dark_background")
```

```python
data = pd.read_parquet("data.parquet")
data.columns
```

```python
t = np.arange(len(data))/1e3
fig,axs = plt.subplots(nrows=3,sharex=True,figsize=(15,15))
ax = axs[0]
ax.plot(t,np.vstack(data["MountEncoders"])[:,:4].mean(1));
ax.grid()
ax = axs[1]
ax.plot(t, np.vstack(data["MountEncoders"])[:, 4:10].mean(1))
ax.grid()
ax = axs[2]
ax.plot(t, np.vstack(data["MountEncoders"])[:, 10:].mean(1))
ax.grid()
```
