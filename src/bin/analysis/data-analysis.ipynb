```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("dark_background")
```

```python
data = pd.read_parquet("data.parquet")
data.columns
```

```python
m1_loads = np.vstack(data["M1Loads"]).reshape(-1,7,6)
m1_loads.shape
```

```python
fig, axs = plt.subplots(ncols=2,nrows=7,figsize=(15,25),sharex=True)
t = np.arange(m1_loads.shape[0])/8e3
for k in range(7):
    axs[k,0].plot(t,m1_loads[:,k,:3]);
    axs[k,1].plot(t,m1_loads[:,k,3:]);
    axs[k,0].grid()
    axs[k,1].grid()
    axs[k,0].set_ylabel("Forces [N]")
    axs[k,1].set_ylabel("Moment [N.m]")
    axs[k,0].set_title(f"Segment #{k+1}")
    axs[k,1].set_title(f"Segment #{k+1}")
    axs[k,1].legend(list("XYZ"))
axs[-1,0].set_xlabel("Time [s]")
axs[-1,1].set_xlabel("Time [s]");
```
