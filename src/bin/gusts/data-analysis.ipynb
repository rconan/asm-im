```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("dark_background")
```

```python
data = pd.read_parquet("data.parquet")
data.columns
```

```python
fig, ax = plt.subplots(figsize=(12,5))
t = np.arange(len(data))/8e3
wind_speeds = [17,12,7,2]
for wind_speed in wind_speeds:
    tt = np.vstack(data[f"Jitter{wind_speed}ms"])*180*3600e3/np.pi
    ax.plot(t,np.hypot(*np.hsplit(tt,2)))
    #ax.plot(t,tt)
ax.set_xlabel("Time [s]")
ax.set_ylabel("Image Jitter RSS [mas]")
ax.grid(alpha=0.5)
ax.legend(wind_speeds)
ax.set_xlim([0,3])
```
